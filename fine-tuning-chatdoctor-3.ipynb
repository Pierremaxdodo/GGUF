```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "PierreMaxime/llama-3-8b-chat-doctor-merged"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16).to("cuda")

```

```python
messages = [
    {
        "role": "user",
        "content": "Hello doctor, I have a bad headache. How do I get rid of it?"
    }
]

def apply_chat_template(messages, tokenize=False, add_generation_prompt=True):
    chat_history = ""
    for msg in messages:
        role = msg["role"]
        content = msg["content"]
        chat_history += f"{role}: {content}\n"
    if add_generation_prompt:
        chat_history += "assistant: "
    return chat_history

prompt = apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True).to("cuda")

outputs = model.generate(**inputs, max_length=250, num_return_sequences=1)

text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(text.split("assistant:")[1])
```
